Design a class called Shape that can represent various geometric shapes. Implement subclasses for specific
shapes like Circle, Rectangle, and RightTriangle. Each subclass should have properties of that shape (e.g..
Width, height for a rectangle) and methods for calculating the area and perimeter of the shape. Create
instances of these shapes and calculate their areas and perimeters.
  - class Shape {

}
class Circle extends Shape {
    constructor(r) {
        super();
        this.radius = r;
    }
    areaOfCircle(){
        let area = Math.PI * this.radius * this.radius;
        console.log(`area of Circle is : ${area}`);
        return area ;
    }
}
class Rectangle extends Shape {
    constructor(h,w) {
        super();
        this.height = h;
        this.width = w;
    }
    areaOfRectangle(){
        let area = this.height * this.width;
        console.log(`area of Rectangle is : ${area}`);
        return area;
    }
}
class RightTriangle extends Shape {
    constructor(b,h) {
        super();
        this.base = b;
        this.height = h;
    }
    areaOfRightTriangle(){
        let area = 0.5 * this.height * this.base;
        console.log(`area of RightTriangle is : ${area}`);
        return area ;
    }
}

Problem Statement 7: Implement a Library System

Create a class called Library that can manage a collection of books. Each book should be represented as an
instance of a Book class (with properties like title, author, and availability). Implement methods in the Library
class to add books, check out books, and return books. Keep track of the books available.
    - class Book{
    constructor(title, author) {
    this.title = title;
    this.author = author;
    this.isAvailable = true; // default to available
  }
}
class Library{
    constructor(){
        this.books = [];
    }
    addBook(book){
        this.books.push(book);
        console.log(`Added: "${book.title}" by ${book.author}`);
    }
    
  // Check out a book by title
  checkoutBook(title) {
    const book = this.books.find(b => b.title === title);
    if (book) {
      if (book.isAvailable) {
        book.isAvailable = false;
        console.log(`You have checked out "${book.title}"`);
      } else {
        console.log(`"${book.title}" is already checked out.`);
      }
    } else {
      console.log(`Book titled "${title}" not found.`);
    }
  }
  returnBook(title) {
    const book = this.books.find(b => b.title === title);
    if (book) {
      if (!book.isAvailable) {
        book.isAvailable = true;
        console.log(`You have returned "${book.title}".`);
      } else {
        console.log(`"${book.title}" was not checked out.`);
      }
    } else {
      console.log(`Book titled "${title}" not found.`);
    }
  }
  // Display all books and their status
  displayBooks() {
    if (this.books.length === 0) {
      console.log("Library is empty.");
    } else {
      console.log("Books in Library:");
      this.books.forEach((book, index) => {
        const status = book.isAvailable ? "Available" : "Checked Out";
        console.log(`${index + 1}. "${book.title}" by ${book.author} - ${status}`);
      });
    }
}
}

const myLibrary = new Library();

const book1 = new Book("Ramayana", "Valmiki");
const book2 = new Book("Mahabharata", "Vyasa");

myLibrary.addBook(book1);
myLibrary.addBook(book2);

myLibrary.displayBooks();

myLibrary.checkoutBook("Ramayana");
myLibrary.displayBooks();

myLibrary.returnBook("Ramayana");
myLibrary.displayBooks();
